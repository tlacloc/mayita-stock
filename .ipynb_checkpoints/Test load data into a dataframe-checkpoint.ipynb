# Liberías utilizadas

```python
import pandas as pd
import json
```

# Abre uno de los archivos recopliados

```python
with open('datos/details-joyeria-0.json') as json_file:
    joyeriaProducts = json.load(json_file)
```

En un mundo facil y sencillo ya tendrían la información sobre la categoría y subcategoría; esto es con el fin de poder hacer una estadista más cómoda, aunque realmente no estoy del todo seguro si fuese la mejor idea; ya que se manejaría una cantidad de datos absurda y terminaremos dividiendo la base de datos por categorías y subcategorías de todos modos, ya que las cotizaciones se realizan por el tipo de producto que es y el precio que este cuesta (más el costo del envio)

# Clean json to make dataframe

```python
#importantKeys = ['id', 'name', 'price (MXN)', 'weight (g)', 'stock', 'category', 'subcategory', 'url', 'img_url']
generalInfo = []

for product in joyeriaProducts:
    # algunos productos tienen subproductos de venta
    for subproduct in product['data']['skuTieredPrices']:
        tmp = {}
        tmp.update({'id' : product['item id'],
                    'name': product['product name'][:-10],
                    'price (MXN)' : product['data']['currencyRate'],
                    'weight (g)' : subproduct['weight'],
                    'stock' : subproduct['inventory'],
                    'category' : product['category'],
                    'subcategory' : product['subcategory'],
                    'url' : product['link']
                   })
        generalInfo.append(tmp)
        
        
df = pd.DataFrame(generalInfo)               
df
```
